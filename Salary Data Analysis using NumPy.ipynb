## Mini Project: Salary Data Analysis using NumPy

```python
import numpy as np

salary = np.array([15000, 22000, 18000, 30000, 25000])
```

**Total Salary**

```python
print("Total Salary = ",salary.sum())
```

**Average Salary**

```python
avrg_salary = salary.mean()
print("Average Salary = ",avrg_salary)
```

**Highest and lowest Salary**

```python
print("Highest Salary = ",salary.max())
print("Lowest Salary = ",salary.min())
```

**Salary Above Average**

```python
print("Above Average Salary = ",salary[salary >avrg_salary])
```

**salary greater the 20000**

```python
get = salary[salary > 20000]
print("salary greater 20000",get)
```
